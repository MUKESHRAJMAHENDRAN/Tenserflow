## @tf function

```python
def function(x,y):
  return x*2 + y*3
 
x = tf.constant(np.arange(1,10,2))
y = tf.constant(np.arange(1,10,2))

function(x,y)
```

```python
@tf.function
def function(x,y):
  return x*2 + y*3

function(x,y)
```

Much of the difference happens behind the scenes. One of the main ones being potential code speed-ups where possible.
